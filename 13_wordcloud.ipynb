워드클라우드 작성 순서

```python
# 워드클라우드로 작성할 말뭉치를 읽어온다.
text = open('./data/alice.txt').read()
print(text)
```

```python
from PIL import Image           # 이미지에서 마스킹 데이터를 뽑아낸다.
from wordcloud import STOPWORDS # 불용어 => 워드클라우드로 작성할 단어에서 제외할 단어
```

```python
# 워드클라우드 모양으로 사용할 이미지를 읽어온다.
alice_mask = np.array(Image.open('./data/alice_mask.png'))
alice_mask
```

```python
stopwords = set(STOPWORDS)
print(type(stopwords))
stopwords
```

```python
stopwords.add('said') # 불용어를 추가한다.
stopwords
```

```python
stopwords.remove('said') # 불용어를 삭제한다.
stopwords
```

```python
# 워드클라우드 설정 및 생성을 한다.
# background_color 속성으로 배경색을 지정한다.
# stopwords 속성으로 불용어를 지정한다.
# mask 속성으로 마스크 이미지를 지정한다.
wc = WordCloud(background_color = 'white', max_words = 2000, stopwords = stopwords, mask = alice_mask) # 워드클라우드 설정
wordcloud = wc.generate(text)
wordcloud.words_
```

```python
# 워드클라우드를 표시한다.
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear') # interpolation(보간법)
plt.axis('off') # 그래프 눈금을 숨긴다.
plt.show()
```

```python
import random
```

```python
# 워드클라우드를 그레이톤으로 표시하는 함수를 만든다. => recolor() 함수의 color_func 속성에 지정한다.
# word => 워드클라우드로 표시할 단어
# font_size => 워드클라우드로 표시할 단어의 크기
# position => 워드클라우드로 표시할 단어의 위치
# orientation => 워드클라우드로 표시할 단어의 방향
# **hwargd => 워드클라우드로 표시할 단어에 적용될 글꼴의 경로
def gray_color_func(word, font_size, position, orientation, random_state=None, **hwargd):
    # print(word, font_size, position, orientation, hwargd)
    # hsl: hue, saturation, lightness
    return 'hsl(0, 0%%, %d%%)' % random.randint(0, 100)
```

```python
plt.figure(figsize=(15, 10))
# recolor() 함수에 그레이톤으로 표시하는 함수와 random_state를 넘겨준다.
plt.imshow(wordcloud.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear') 
plt.axis('off') 
plt.show()
```

```python
import nltk 
from konlpy.corpus import kolaw
from konlpy.tag import Twitter
t = Twitter()
```

```python
# konlpy 패키지는 대한민국 헌법 말뭉치 kolaw와 국회법안 말뭉치인 kobill을 제공한다.
# constitution.txt 파일은 kolaw에 포함된 파일로 데이터 파일 없이 파일 이름만 지정하면 읽어올 수 있다. 
ko_con_text = kolaw.open('constitution.txt').read()
print(ko_con_text)
```

```python
# 한글 데이터 형태소 분석을 실행한다.
tokens_ko = t.nouns(ko_con_text) # 명사 뽑기
tokens_ko
```

```python
# 불용어를 정의한다.
stop_words = ['제', '년', '월', '일', '조', '수', '때', '그', '이', '바', '및', '정', '관', '위', '장', '안', '거나', '이상', '다만', '인']
# 형태소 분석 결과에서 불용어를 제거한다.
# 형태소 분석 결과(tokens_ko)의 각 단어가 each_word에 저장되며 반복한다.
# 이때, each_word에 저장된 단어가 stop_words에 포함되지 않는 단어만 리스트로 만들어 형태소 분석 결과로 저장한다.
tokens_ko = [each_word for each_word in tokens_ko if each_word not in stop_words]
tokens_ko
```

```python
# 형태소 분석 결과 단어별 출현 횟수를 계산한다. 
ko = nltk.Text(tokens_ko)
ko.vocab().most_common(50)
```

```python
# 단어별 출현 횟수가 많은 순서대로 워드클라우드로 표현할 단어를 추출하고 딕셔너리 형태로 만든다.
data = ko.vocab().most_common(500)
tmp_data = dict(data)
tmp_data
```

```python
# 딕셔너리에 저장된 데이터는 generate_from_frequencies() 함수를 사용해서 워드클라우드를 만든다.
wordcloud = WordCloud(
    font_path='C:\\Windows\\Fonts\\NanumGothicCoding.ttf',
    background_color='black',
    relative_scaling=0.5    
).generate_from_frequencies(tmp_data)
```

```python
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis('off') 
plt.show()
```
