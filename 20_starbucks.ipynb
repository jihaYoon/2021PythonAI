```python
import warnings
warnings.filterwarnings('ignore')
import requests
```

시도 코드 크롤링

```python
targetSite = 'https://www.starbucks.co.kr/store/getSidoList.do'
# Request URL 을 서버에 요청하는 방식인 Request Method가 POST 방식이므로
# requests 모듈의 post() 메소드로 서버에 정보를 요청한다.
# Request Method가 POST 방식일 경우 서버에 정보를 요청할 때 넘겨주는 데이터가 있으므로 
# Headers 맨 아래의 Form data 내용을 data 속성에 딕셔너리 형태로 전달해야 한다.
request = requests.post(targetSite, data={
    'rndCod': 'XGW4SNMGHM'
})
print(request)
print(type(request.text))
print(request.text)
```

```python
# json 모듈의 loads() 함수로 크롤링한 json 타입의 문자열을 파이썬에서 처리하기 위해서 딕셔너리 타입으로 변환한다.
import json
star_json = json.loads(request.text)
print(type(star_json))
print(star_json)
```

```python
# requests 모듈의 json() 함수로 크롤링한 json 타입의 문자열을 파이썬에서 처리하기 위해서 딕셔너리 타입으로 변환한다
sidoList = request.json()
print(type(sidoList))
```

```python
print(len(sidoList))
print(len(sidoList['list']))
```

```python
print(sidoList['list'][0])
```

```python
# 시도 코드의 개수만큼 반복하며 시도코드['sido_cd']를 key로 하고 시도이름('sido_nm')을 value로 하는 딕셔너리를 만든다.
sido_List = {}
for sido in sidoList['list']:
    # print('{}: {}'.format(sido['sido_cd'], sido['sido_nm']))
    sido_List[sido['sido_cd']] = sido['sido_nm']
print(sido_List)
```

구군 코드 크롤링

```python
# 크롤링할 시도 코드를 입력받는다.
sido_cd = input('시군코드를 입력하세요: ')
targetSite = 'https://www.starbucks.co.kr/store/getGugunList.do'
request = requests.post(targetSite, data={
    'sido_cd': sido_cd,
    'rndCod': 'KO8YCH9TS6'
})
# print(request)
gugunList = request.json()
# print(type(gugunList))

# 구군코드의 개수만큼 반복하며 구군코드('gugun_cd')를 key로 하구 구군이름('gugun_nm')을 value로 하는 딕셔너리를 만든다.
gugun_List = {}
for gugun in gugunList['list']:
    gugun_List[gugun['gugun_cd']] = gugun['gugun_nm']
print(gugun_List)
```

시도/구군의 지점별 크롤링

```python
targetSite = 'https://www.starbucks.co.kr/store/getStore.do?r=ZC765RZQRL'
request = requests.post(targetSite, data={
    'ins_lat': 37.2834304 ,    # 경도
    'ins_lng': 127.0480896 ,   # 위도
    'p_sido_cd': '' ,          # 시도코드
    'p_gugun_cd': '0815' ,         # 구군코드
    'in_biz_cd': '' ,          # ? 
    'iend': 100 ,              # 서버에서 응답하는 매장의 최대 개수  
    'set_date': ''             # ?
    # 'in_biz_cd', 'set_date'를 넘겨주지 않으면 서버에서 데이터를 받아오지 못한다.(필수)
})
storeList = request.json()
# print(len(storeList['list']))

count = 0
dt_count =0 
for store in storeList['list']:
    count += 1
    print('{0:4d}. {1}: {2}({3}, {4})'.format(count, store['s_name'], store['doro_address'], store['lat'], store['lot']))
    
    if store['s_name'].find('DT') >= 0 :
        dt_count += 1
        
print('DT 매장 개수 : {}'.format(dt_count))
```
