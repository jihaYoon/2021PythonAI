```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as df
```

```python
print(df.now)
```

```python
targetSite = 'https://music.bugs.co.kr/chart?wl_ref=M_left_02_01'
request = requests.get(targetSite)
# print(request)
html = request.text
soup = BeautifulSoup(html,'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p',{'class':'title'})
# print(len(titles))

for i in range (len(titles)):
    # print('{0:3d}위 {1}'.format((i+1), titles[i].text.strip())) # strip(): 공백을 없애는 메소드
    # print('{0:3d}위 {1}'.format((i+1), titles[i].text.split('\n')[1]))
    print('{0:3d}위 {1}'.format((i+1), titles[i].text.strip().split('\n')[0]))
```

```python
# 아티스트 크롤링
artists = soup.findAll('p',{'class':'artist'})
# print(artists)
# print(len(artists))

for i in range (len(artists)):
    # print(artists[i].text.strip())
    #print(artists[i].text.split('\n')
    print('{0:3d}위 {1}'.format(i+1, artists[i].text.strip().split('\n')[0]))
```

```python
print(df.now())

for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{0:3d}위 {1} - {2}'.format(i+1, title, artist))
```

```python
# 크롤링한 결과를 텍스트 파일로 저장한다.
file = open('./output/bugsTOP100.txt','w')
file.write('{} 현재 Bugs 뮤직 실시간 TOP 100\n'.format(df.now()))

for i in range(100):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    file.write('{0:3d}위 {1} - {2}\n'.format(i+1, title, artist))

file.close()
print('bugsTOP100.txt 파일로 쓰기 완료')
```
