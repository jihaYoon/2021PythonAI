```python
park['공원면적'].head()
```

```python
# park['공원면적'] 열의 데이터가 처음부터 차례대로 lambda 변수 x에 저장되면서 뒤의 수식을 실행한 결과를 적용시킨다.
park['공원면적비율'] = park['공원면적'].apply(lambda x: np.sqrt(x) * 0.01)

    # for i in park['공원면적']:
    #    print(np.sqrt(x) * 0.01)

park.head()
```

```python
# 도로명 주소가 NaN인 데이터의 개수 => 지번 주소만 입력된 데이터의 개수
# 도로명 주소만 입력되고 지번 주소가 입력되지 않은 데이터는 도로명 주소 제도가 시행되고 난 후 조성된 공원이다.
park['소재지도로명주소'].isnull().sum()
```

```python
# 도로명 주소는 입력되지 않고 지번 주소만 입력된 데이터
park[(park['소재지도로명주소'].isnull()) & (park['소재지지번주소'].notnull())]
```

```python
# 지번주소가 NaN인 데이터의 개수 => 도로명 주소만 입력된 데이터
park[(park['소재지도로명주소'].notnull()) & (park['소재지지번주소'].isnull())]
```

```python
# 도로명 주소와 지번주소 모두 입력된 데이터
park[(park['소재지도로명주소'].notnull()) & (park['소재지지번주소'].notnull())]
```

```python
# 도로명 주소가 NaN인 데이터를 지번 주소로 채우기
park['소재지도로명주소'].fillna(park['소재지지번주소'], inplace=True)
park
```

```python
park['소재지도로명주소'].isnull().sum()
```

```python
# split() 함수 실행시 expand=True 옵션을 지정하면 데이터프레임으로 분리되어 인덱싱과 슬라이싱을 사용할 수 있다.
park['소재지도로명주소'].str.split(' ', expand=True)
```

```python
# 도로명 주소에서 '시도'만 추출해서 '시도' 열을 만들어 park 데이터프레임에 추가한다.
park['시도'] = park['소재지도로명주소'].str.split(' ', expand=True)[0]
park['구군'] = park['소재지도로명주소'].str.split(' ', expand=True)[1]
park['읍면동'] = park['소재지도로명주소'].str.split(' ', expand=True)[2]

park.head()
```

```python
# 위도, 경도 이상값 추출하기
park_error = park[(park['위도'] < 32) | (park['경도'] > 132)]
park_error
```

```python
# 올바른 데이터
park_ok = park[(park['위도'] >= 32) & (park['경도'] < 132)]
park_ok
```

```python
park_ok['시도'].value_counts()
```

```python
# '시도'에 '강원'으로 입력된 데이터를 '강원도'로 수정한다.
park_ok['시도'][park['시도'] == '강원'] = '강원도'
park_ok['시도'].value_counts()
```

시도별 공원 비율

```python
park['시도'][park['시도'] == '강원'] = '강원도'
```

```python
park['시도'].value_counts()
```

```python
# '시도'별 개수 데이터를 계산한다.
# value_counts() : 별도의 옵션을 지정하지 않으면 지정된 데이터 개수를 계산해서 내림차순으로 정렬한다.
# ascending=True 옵션을 지정하면 오름차순 정렬된다.

park_do = pd.DataFrame(park['시도'].value_counts(ascending=True))
park_do
```

```python
# value_counts() : 별도의 옵션을 지정하지 않으면 데이터 개수를 계산하고 
# normalize=True 옵션을 지정하면 전체 개수에서 차지하는 비율을 계산한다.

park_normalize = pd.DataFrame(park['시도'].value_counts(normalize=True))
park_normalize
```

```python
# 시도별 개수 데이터와 개수의 비율 데이터를 병합하고 인덱스를 다시 부여한다.
park_sido = park_do.merge(park_normalize, left_index=True, right_index=True).reset_index()

# 병합된 데이터의 열 이름 변경
park_sido.columns = ['시도', '개수', '비율']
park_sido
```

```python
ggplot(park_sido, aes(x='시도', y='개수'))\
    + geom_bar(stat='identity', fill='green') \
    + theme(text=element_text(family='NanumGothic'), figure_size=(12, 8), axis_text_x=element_text(rotation=90))
```

```python
ggplot(park_sido, aes(x='시도', y='개수'))\
    + geom_bar(stat='identity', fill='green') \
    + coord_flip() \
    + theme(text=element_text(family='NanumGothic'), figure_size=(12, 8))
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=park_sido, x='개수', y='시도')
```

```python
plt.figure(figsize=(12,10))
plt.xticks(rotation=45)
sns.barplot(data=park_sido, x='시도', y='개수')
```

```python
from pandas.api.types import CategoricalDtype
from plotnine.data import mpg
```

```python
(ggplot(mpg) 
 + aes(x='manufacturer')
 + geom_bar(size=20)
 + coord_flip()
 + labs(y='Count', x='Manufacturer', title='Number of Cars by Make')
)
```

```python
# Note that value_counts() is already sorted
manufacturer_list = mpg['manufacturer'].value_counts().index.tolist()
manufacturer_cat = pd.Categorical(mpg['manufacturer'], categories=manufacturer_list)

# assign to a new column in the DataFrame
mpg = mpg.assign(manufacturer_cat = manufacturer_cat)

(ggplot(mpg)
 + aes(x='manufacturer_cat')
 + geom_bar(size=20)
 + coord_flip()
 + labs(y='Count', x='Manufacturer', title='Number of Cars by Make')
)
```

```python
park_list = park['시도'].value_counts().index.tolist()
park_cat = pd.Categorical(park['시도'], categories=park_list)


park = park.assign(park_cat = park_cat)

ggplot(park, aes(x='park_cat')) \
 + geom_bar(size=20) \
 + coord_flip() \
 + labs(y='개수', x='시도', title='시도별 공원 분포수')\
 + theme(text=element_text(family='NanumGothic'), figure_size=(12, 8))

```
